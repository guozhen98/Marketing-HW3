```python
import pandas as pd
import numpy as np
```

```python
#取出convert_TF为True的行
attribution = pd.read_csv('attribution_allocation_student_data.csv')
attribution=attribution[attribution['convert_TF']==True]
attribution
```

```python
#First interaction
attributionMatrix=pd.DataFrame(columns=['email', 'social', 'display','paid_search','referral'],index=['first1','first2','first3','last1','last2','last3','linear1','linear2','linear3'])
#tier1
a = attribution[attribution['tier']==1]['touch1'].value_counts()
attributionMatrix.at['first1','email']  = a['email']
attributionMatrix.at['first1','social'] = a['social']
attributionMatrix.at['first1','display'] = a['display']
attributionMatrix.at['first1','paid_search'] = a['paid_search']
attributionMatrix.at['first1','referral'] = a['referral']
#tier2
a = attribution[attribution['tier']==2]['touch1'].value_counts()
attributionMatrix.at['first2','email']  = a['email']
attributionMatrix.at['first2','social'] = a['social']
attributionMatrix.at['first2','display'] = a['display']
attributionMatrix.at['first2','paid_search'] = a['paid_search']
attributionMatrix.at['first2','referral'] = a['referral']
#tier3
a = attribution[attribution['tier']==3]['touch1'].value_counts()
attributionMatrix.at['first3','email']  = a['email']
attributionMatrix.at['first3','social'] = a['social']
attributionMatrix.at['first3','display'] = a['display']
attributionMatrix.at['first3','paid_search'] = a['paid_search']
attributionMatrix.at['first3','referral'] = a['referral']

attributionMatrix
```

```python
#把touch过程变成数组
list1=attribution[['touch1','touch2','touch3','touch4','touch5']]
touchList = []
for index, row in list1.iterrows():
    a=row.tolist()
    a = [a_ for a_ in a if a_ == a_]
    touchList.append(a)
touchList
```

```python
# Last interaction
lastIn = []
for i in touchList:
    #print(i[len(i)-1])
    lastIn.append(i[len(i)-1])
attribution['lastTouch']=lastIn    

#tier1
a = attribution[attribution['tier']==1]['lastTouch'].value_counts()
attributionMatrix.at['last1','email']  = a['email']
attributionMatrix.at['last1','social'] = a['social']
attributionMatrix.at['last1','display'] = a['display']
attributionMatrix.at['last1','paid_search'] = a['paid_search']
attributionMatrix.at['last1','referral'] = a['referral']
#tier2
a = attribution[attribution['tier']==2]['lastTouch'].value_counts()
attributionMatrix.at['last2','email']  = a['email']
attributionMatrix.at['last2','social'] = a['social']
attributionMatrix.at['last2','display'] = a['display']
attributionMatrix.at['last2','paid_search'] = a['paid_search']
attributionMatrix.at['last2','referral'] = a['referral']
#tier3
a = attribution[attribution['tier']==3]['lastTouch'].value_counts()
attributionMatrix.at['last3','email']  = a['email']
attributionMatrix.at['last3','social'] = a['social']
attributionMatrix.at['last3','display'] = a['display']
attributionMatrix.at['last3','paid_search'] = a['paid_search']
attributionMatrix.at['last3','referral'] = a['referral']

attributionMatrix
```
